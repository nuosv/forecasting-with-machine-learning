In this notebook, we will predict energy demand using simple forecasting models, to obtain a benchmark against which we can test our machine learning models later on.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.metrics import mean_squared_error

from skforecast.ForecasterBaseline import ForecasterEquivalentDate
from skforecast.model_selection import backtesting_forecaster
```

# Load data

We will use the electricity demand dataset found [here](https://github.com/tidyverts/tsibbledata/tree/master/data-raw/vic_elec/VIC2015).

**Citation:**

Godahewa, Rakshitha, Bergmeir, Christoph, Webb, Geoff, Hyndman, Rob, & Montero-Manso, Pablo. (2021). Australian Electricity Demand Dataset (Version 1) [Data set]. Zenodo. https://doi.org/10.5281/zenodo.4659727

**Description of data:**

A description of the data can be found [here](https://rdrr.io/cran/tsibbledata/man/vic_elec.html). The data contains electricity demand in Victoria, Australia, at 30 minute intervals over a period of 12 years, from 2002 to early 2015. There is also the temperature in Melbourne at 30 minute intervals and public holiday dates.

```python
# Electricity demand.
url = "https://raw.githubusercontent.com/tidyverts/tsibbledata/master/data-raw/vic_elec/VIC2015/demand.csv"
df = pd.read_csv(url)

df.drop(columns=["Industrial"], inplace=True)

# Convert the integer Date to an actual date with datetime type
df["date"] = df["Date"].apply(
    lambda x: pd.Timestamp("1899-12-30") + pd.Timedelta(x, unit="days")
)

# Create a timestamp from the integer Period representing 30 minute intervals
df["date_time"] = df["date"] + pd.to_timedelta((df["Period"] - 1) * 30, unit="m")

df.dropna(inplace=True)

# Rename columns
df = df[["date_time", "OperationalLessIndustrial"]]

df.columns = ["date_time", "demand"]

# Resample to hourly
df = (
    df.set_index("date_time")
    .resample("H")
    .agg({"demand": "sum"})
)

df.head()
```

```python
df.tail()
```

```python
df.index.min(), df.index.max()
```

```python
# Split into train and test
# We leave the last February in the test set

end_train = '2014-12-31 23:59:59'
X_train = df.loc[:end_train]
X_test  = df.loc[end_train:]
```

```python
fig, ax=plt.subplots(figsize=(7, 3))
X_train.plot(ax=ax, label='train')
X_test.plot(ax=ax, label='test')
ax.set_title('Hourly energy consumption.')
ax.legend(["train", "test"])
ax.legend();
```

```python
# Let's zoom in to see a bit more detail:

fig, ax=plt.subplots(figsize=(7, 3))
X_train.tail(2000).plot(ax=ax)
X_test.plot(ax=ax)
ax.set_title('Hourly energy consumption.')
ax.legend(["train", "test"])
plt.show()
```

## Naive forecasting

We predict the next value as its immediate value in the past.

```python
# The Baseline forecasters from skforecast take series as input

pd.Series(X_train["demand"])
```

```python
forecaster = ForecasterEquivalentDate(
    offset=1, # naive forecasting
    forecaster_id = "naive"
)

# takes a copy of the last relevant window in the train set
# to predict the first point in the horizon in the test set

forecaster.fit(
    y=pd.Series(X_train["demand"]), # input needs to be a series
)

forecaster
```

```python
# we want to forecast the next 24 hs

predictions = forecaster.predict(steps=24)

# as expected, the forecast is a constant

predictions
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
X_train.tail(100).plot(ax=ax, label='train')
X_test.head(24).plot(ax=ax, label='train')
predictions.plot(ax=ax, label='predictions')
plt.title("Lasso forecasting")
ax.legend(bbox_to_anchor=(1.3, 1.0));
```

```python
# we want to forecast 24 hs, starting 2nd January

# In our scenario, we have all data available up to the hr before 
# the forecast. So we have this data available:

last_data = X_test.loc[:'2015-01-01 23:59:59']
last_data = pd.Series(last_data["demand"])

last_data
```

```python
predictions = forecaster.predict(
    steps=24,
    last_window=last_data

)

# as expected, the forecast is a constant

predictions
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
X_train.tail(100).plot(ax=ax, label='train')
X_test.loc['2015-01-02 00:00:00':'2015-01-02 23:59:59'].plot(ax=ax, label='train')
predictions.plot(ax=ax, label='predictions')
plt.title("Lasso forecasting")
ax.legend(bbox_to_anchor=(1.3, 1.0));
```

## Seasonal naive forecasting

We predict the next value as its value a season ago. For our case, we know that the time series shows weekly seasonality.

```python
# Finally, we can use the same hour, but in the previous week

freq = "6d"
df["seasonal"] = df["demand"].shift(freq=freq)

df.head()
```

## Moving average

We predict the next value as the average of the last 3 values.

```python
# We'll use a 3H window size.

result = (
    df["demand"]
    .rolling(window=3) # Pick window size.
    .agg(["mean"]) # Pick statistics.
    .shift(freq="1H") # Lag by 1 hour to avoid data leakage.
)  

result.columns = ["moving_ave"]

result
```

```python
# add features to main dataframe

df = df.merge(result, how="left", left_index=True, right_index=True)

df.head()
```

## Historical

We predict future values as the mean of all values in the past.

```python
result = (
    df["demand"]
    .expanding() # all values in the past.
    .agg(["mean"]) # Pick statistics.
    .shift(freq="1H") # Lag by 1 hour to avoid data leakage.
)  

result.columns = ["historical_ave"]

# add features to main dataframe
df = df.merge(result, how="left", left_index=True, right_index=True)

df.head(30)
```

```python
# Let's drop all those rows where we don't have forecasts

df.dropna(inplace=True)

df.head()
```

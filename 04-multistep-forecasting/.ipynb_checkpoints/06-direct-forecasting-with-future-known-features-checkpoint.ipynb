# Direct Forecasting with feature pipelines

[Forecasting with Machine Learning - Course](https://www.trainindata.com/p/forecasting-with-machine-learning)

In this notebook, we carry out **direct forecasting** to predict multiple steps into the future by using a gradient boosting machine for regression.

SKForecast produces lags out of the box. But we can combine SKForecast with Feature-engine and other libraries to create more features.

```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

from skforecast.ForecasterAutoregDirect import ForecasterAutoregDirect

from feature_engine.datetime import DatetimeFeatures
from feature_engine.imputation import MeanMedianImputer
from feature_engine.timeseries.forecasting import WindowFeatures
```

# Load data

We will use the electricity demand dataset found [here](https://github.com/tidyverts/tsibbledata/tree/master/data-raw/vic_elec/VIC2015).

**Citation:**

Godahewa, Rakshitha, Bergmeir, Christoph, Webb, Geoff, Hyndman, Rob, & Montero-Manso, Pablo. (2021). Australian Electricity Demand Dataset (Version 1) [Data set]. Zenodo. https://doi.org/10.5281/zenodo.4659727

**Description of data:**

A description of the data can be found [here](https://rdrr.io/cran/tsibbledata/man/vic_elec.html). The data contains electricity demand in Victoria, Australia, at 30 minute intervals over a period of 12 years, from 2002 to early 2015. There is also the temperature in Melbourne at 30 minute intervals and public holiday dates.

```python
# Electricity demand.
url = "https://raw.githubusercontent.com/tidyverts/tsibbledata/master/data-raw/vic_elec/VIC2015/demand.csv"
df = pd.read_csv(url)

df.drop(columns=["Industrial"], inplace=True)

# Convert the integer Date to an actual date with datetime type
df["date"] = df["Date"].apply(
    lambda x: pd.Timestamp("1899-12-30") + pd.Timedelta(x, unit="days")
)

# Create a timestamp from the integer Period representing 30 minute intervals
df["date_time"] = df["date"] + \
    pd.to_timedelta((df["Period"] - 1) * 30, unit="m")

df.dropna(inplace=True)

# Rename columns
df = df[["date_time", "OperationalLessIndustrial"]]

df.columns = ["date_time", "demand"]

# Resample to hourly
df = (
    df.set_index("date_time")
    .resample("H")
    .agg({"demand": "sum"})
)

df.head()
```

```python
df.tail()
```

```python
# Split into train and test

# We leave the last February in the test set

end_train = '2014-12-31 23:59:59'
X_train = df.loc[:end_train]
X_test  = df.loc[end_train:]
```

## Plot time series

```python
fig, ax = plt.subplots(figsize=(7, 3))
X_train.tail(500).plot(ax=ax)
X_test.head(500).plot(ax=ax)
ax.set_title('Hourly energy consumption.')
ax.legend(["train", "test"])
plt.show()
```

## Regression model

```python
model = GradientBoostingRegressor(random_state=9, max_depth=3) 
```

## Extracting features

We'll add window features and datetime features.

```python
window_f = WindowFeatures(
    variables="demand",        # the variable over which the windows will be created
    window=[3, 6, 24],         # the size of the windows
    functions=["mean", "std"], # the functions to use to aggregate data
    missing_values="ignore",   # this is important for predictions, more later
    drop_original=True,        # we need to drop the variable, because it is in essence, the target
)

# test transformer
window_f.fit_transform(X_train)
```

WindowFeatures is working as expected. So now, we line it up in a pipeline with the datetime features to create all of this features at the same time.

```python
# We will set the datetime features to obtain the features
# from the index instead of from a column.

datetime_f = DatetimeFeatures(
    variables="index",   # to extract features from the index
    features_to_extract=["month", "day_of_week", "hour"]
)
```

```python
# all steps in a pipeline

pipe = Pipeline([
    ("datetime_f", datetime_f),   # the datetime features
    ("window_f", window_f),       # the window features
    ("impute", MeanMedianImputer()) # to impute the nan created by the window features
])

# test pipeline
pipe.fit_transform(X_train)
```

## Direct forecaster

```python
forecaster = ForecasterAutoregDirect(
    regressor=model,          # the machine learning model
    lags=24,                  # the lag features to create
    transformer_exog=pipe,    # to get the datetime features
    steps=24,
    forecaster_id="direct",
)
```

```python
forecaster.fit(
    y=X_train["demand"], # the series for the lags
    exog=X_train,        # the datetime for the datetime features
)

forecaster
```

## Input features to regression

```python
forecaster.create_train_X_y(
    y=X_train["demand"],
    exog=X_train,
)
```

Note that we now have multiple targets, one per step in the horizon.

## Predict future data

We'll start by predicting the 24 hs right after the train set ends.

We need to create a window with the forecasting horizon.

```python
forecast_point = '2015-01-01 00:00:00'
forecast_end = '2015-01-01 23:59:59'

index = pd.date_range(
    start=forecast_point,
    end=forecast_end,
    freq="1H",
)

index
```

```python
f_horizon = pd.DataFrame(
    columns=["demand"], index=index)

f_horizon
```

```python
preds = forecaster.predict(
    steps=24,
    exog=f_horizon,
)

preds
```

```python
# Plot the forecast vs the actual

fig, ax = plt.subplots(figsize=(6, 3))
X_train.tail(100)["demand"].plot(ax=ax, label='train')
X_test.head(24)["demand"].plot(ax=ax, label='test')
preds.plot(ax=ax, label='predictions')
plt.title("GBM recursive forecasting")
ax.legend(bbox_to_anchor=(1.3, 1.0));
```

```python
# Prediction error

error_mse = mean_squared_error(
                y_true = X_test["demand"].head(24),
                y_pred = preds
            )

print(f"Test error (mse): {error_mse}")
```

## Predict any time point in the future

```python
# Say we want to predict energy demand for 1st of February

forecast_start = '2015-02-01 00:00:00'

# we need the energy demand up to 144 hs before that point
past_data_available = X_test[:'2015-01-31 23:59:59'].tail(144)

past_data_available.head()
```

```python
# we also need the datetime in that date range
# To make predictions `exog` must start one step ahead of `last_window`.

horizon = X_test['2015-01-31 23:59:59':].head(24)
```

```python
predictions = forecaster.predict(
    steps=24,
    last_window=past_data_available["demand"],
    exog=horizon,
)

predictions.head()
```

```python
# Plot the forecast vs the actual

fig, ax = plt.subplots(figsize=(6, 3))
horizon["demand"].plot(ax=ax, label='actuals')
predictions.plot(ax=ax, label='predictions')
plt.title("Lasso forecasting")
ax.legend(bbox_to_anchor=(1.3, 1.0));
```

```python
# Prediction error

error_mse = mean_squared_error(
                y_true = horizon,
                y_pred = predictions
            )

print(f"Test error (mse): {error_mse}")
```

We see that this model makes a smaller error in this forecasting window respect to the Lasso (previous notebook), but a larger error on the previous forecasting window.
